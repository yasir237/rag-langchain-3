```python
from google.colab import userdata
import os

os.environ["GROQ_API_KEY"]   = userdata.get("GROQ_API_KEY")
```

```python
from langchain_core.prompts import PromptTemplate
from langchain_groq import ChatGroq

chat = ChatGroq(
    model="llama-3.1-8b-instant",
    temperature=0,
    api_key=os.environ["GROQ_API_KEY"]
)
```

```python
template = """
Rol: Sen deneyimli bir yapay zeka ve derin öğrenme uzmanısın.

Görev:
Aşağıdaki kavramı teknik ama anlaşılır şekilde açıkla:

Kavram:
{concept}

Kurallar:
- Maksimum 120 kelime
- Gereksiz tekrar yapma
- Tanım + nasıl çalışır + nerede kullanılır şeklinde açıkla
- Türkçe yaz
- Başlıklar **başlık** şeklinde yaz

Çıktı formatı:
Açıklama:
"""

prompt = PromptTemplate(
    input_variables=["concept"],
    template=template,
)

chain_one = prompt | chat

response1 = chain_one.invoke({"concept": "autoencoder"})
print(response1.content)
```

```python
second_prompt = PromptTemplate(
    input_variables=["ml_concept"],
    template="""
Aşağıdaki metni referans al ve 5 yaşındaki bir çocuğa anlatır gibi açıkla.
{ml_concept}

Kurallar:
- Maksimum 120-150 kelime
- Tekrar eden cümleler yazma
- Günlük hayattan sadece 1 örnek ver
- Basit ama mantıklı bir açıklama yap
- Direkt anlatmaya başlanmalıdır.
- metnin başında merhaba gibi karşılaşma ifadeler olmamalıdır.
- metnin başında ne anlatacağını yazmadan anlatmaya başla.
"""
)

chain_two = second_prompt | chat

response2 = chain_two.invoke({"ml_concept": response1.content})
print(response2.content)
```

```python
third_prompt = PromptTemplate(
    input_variables=["st_concept"],
    template="""
Görev:
Aşağıdaki metni basit bir dille özetle.

- Metin çok temiz. Alakasız bir şey yazılmamalıdır.
- Cevabın başında 'özet' 'kısaca' gibi kelimeler olmamalıdır.
- Özet anlaşılır ve metnin gidişatını açıklamalıdır.
- Özet Türkçe olmalıdır.

Metin:
{st_concept}
"""
)

chain_three = third_prompt | chat

summary = chain_three.invoke({
    "st_concept": response1.content
}).content

response3 = f"""
**Anlatım**
{response2.content}

**Özet**
{summary}
"""

print(response3)
```

```python
from langchain_core.runnables import RunnableLambda, RunnablePassthrough

overall_chain = (
    chain_one
    | RunnableLambda(lambda x: {"ml_concept": x.content})
    | RunnablePassthrough.assign(
        child_explanation=lambda x: chain_two.invoke({"ml_concept": x["ml_concept"]}).content
    )
    | RunnableLambda(lambda x: {
        "st_concept": x["ml_concept"],      # chain_one çıktısı → teknik özet için
        "anlatim": x["child_explanation"]    # chain_two çıktısı → çocuğa anlatım
    })
    | RunnablePassthrough.assign(
        ozet=lambda x: chain_three.invoke({"st_concept": x["st_concept"]}).content
    )
)

response4 = overall_chain.invoke({"concept": "autoencoder"})
```

```python
print("The 1. response: ", response1.content)
```
